# The 'isna' Method

```python
import pandas as pd
pd.options.display.float_format = '{:.1f}'.format
```

```python
ufos_path = 'data/ufo_sightings.csv'
```

```python
ufo_sightings = pd.read_csv(ufos_path)
ufo_sightings.head()
```

```python
 ufo_sightings.info()
```

```python
null_state = ufo_sightings['state'].isna()
null_state.head()
```

```python
ufo_sightings[null_state].head(10)
```

```python
 null_country = ufo_sightings['country'].isna()

either_null = ufo_sightings[null_state | null_country]
either_null.head(10)
```

```python
len(either_null)
```

## Finding Null Values in Rows

```python
any_null = ufo_sightings.isna().any(axis=1)
any_null.head()
```

```python
# index the dataframe
any_null_df = ufo_sightings[any_null]
any_null_df.head(20)
```

```python
len(any_null_df)
```

### finding numbers that can be converted

```python
# Add a column that has a numeric version of one of the columns that can be converted to numeric but is currently assigned object data type
ufo_sightings['duration_seconds_numeric'] = pd.to_numeric(ufo_sightings['duration (seconds)'], errors='coerce')
```

```python
seconds_nan = ufo_sightings[ufo_sightings['duration_seconds_numeric'].isna()]
seconds_nan[['duration (seconds)', 'duration_seconds_numeric']]
```
