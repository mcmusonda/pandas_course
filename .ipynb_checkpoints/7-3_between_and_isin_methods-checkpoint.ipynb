# Chapter 7-3: The between and isin Methods

```python
import pandas as pd
pd.options.display.float_format = '{:.1f}'.format
```

```python
house_path = 'data/kc_house_data.csv'
```

```python
house_sales = pd.read_csv(house_path)
house_sales.head()
```

## Using the between Method

```python
between_6_n_10 = house_sales['bedrooms'].between(6, 10)
between_6_n_10_df = house_sales[between_6_n_10]
```

```python
between_6_n_10_df.sort_values('bedrooms', inplace=True)
between_6_n_10_df.head()
```

```python
between_6_n_10_df.tail()
```

## Using isin Method
Used to select multiple distinct values in a single column

```python
ufos_path = 'data/ufo_sightings.csv'
ufo_sightings = pd.read_csv(ufos_path)
ufo_sightings.head()
```

```python
ufo_sightings['shape'].value_counts()
```

```python
shapes = ufo_sightings['shape'].isin(['cigar', 'cylinder', 'cigar', 'cone'])
filtered_df = ufo_sightings[shapes]
```

```python
filtered_df['shape'].value_counts()
```
