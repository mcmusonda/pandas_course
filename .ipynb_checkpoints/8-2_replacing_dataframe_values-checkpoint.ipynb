# Chapter 8-2: Replacing DataFrame Values

```python
import pandas as pd
pd.options.display.float_format = '{:.1f}'.format
```

```python
ufos_path = 'data/ufo_sightings.csv'
```

```python
ufo_sightings = pd.read_csv(ufos_path)
ufo_sightings.head()
```

```python
ufo_sightings['country'].value_counts()
```

```python
updated_countries = ufo_sightings['country'].replace('de', 'Germany')                          
```

```python
updated_countries.value_counts()
```

```python
old_country_names = ['us', 'ca', 'gb', 'au', 'de']
new_country_names = ['United States', 'Canada', 'Great Britain', 'Australia', 'Germany']

updated_countries = ufo_sightings['country'].replace(old_country_names,new_country_names)
```

```python
updated_countries.value_counts()
```

```python
ufo_sightings['country'] = updated_countries
```
