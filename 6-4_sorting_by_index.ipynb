# Chapter 6-4: Sorting by Index

```python
import pandas as pd
pd.options.display.float_format = '{:.1f}'.format
```

```python
house_path = 'data/kc_house_data.csv'
```

```python
house_sales = pd.read_csv(house_path, index_col='id')
house_sales.head()
```

## Sort rows according to index and change dataFrame

```python
house_sales.sort_index(inplace=True)
house_sales.head()
```

```python
top_1000 = house_sales.iloc[0:1000]
top_1000.head()
```

## Retrieving the last 1000 rows

```python
house_sales.sort_index(inplace=True, ascending=False)
house_sales.head()
```

```python
bottom_1000 = house_sales.iloc[0:1000]
bottom_1000.head()
```

## Comparing summary statistics

```python
top_1000.describe()
```
