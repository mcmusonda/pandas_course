# Chapter 7-7: Negation

```python
import pandas as pd
```

```python
house_path = 'data/kc_house_data.csv'
```

```python
house_sales = pd.read_csv(house_path)
```

```python
greater_than_1_mil = house_sales['price'] > 1000000
one_floor = house_sales['floors'] == 1
combined = greater_than_1_mil & one_floor

short_n_swanky = house_sales[combined]
short_n_swanky.head(10)
```

```python
less_than_eq_1_mil = house_sales['price'] <= 1000000
gt_one_floor = house_sales['floors'] > 1
Not_short_n_swanky = less_than_eq_1_mil | gt_one_floor
```

```python
house_sales[Not_short_n_swanky].head(10)
```

```python
~greater_than_1_mil.head(10)
```

```python
~gt_one_floor.head(10)
```

```python
greater_than_1_mil = house_sales['price'] > 1000000
one_floor = house_sales['floors'] == 1
combined = greater_than_1_mil & one_floor

NOT_short_n_swanky = house_sales[~combined]
NOT_short_n_swanky.head(10) 
```

```python
NOT_short_n_swanky[NOT_short_n_swanky['price'] > 1000000]
```

```python
price = house_sales['price'] <= 300000
basement = house_sales['sqft_basement'] > 0
floors = house_sales['floors'] > 1

multi_level = house_sales[price & (basement | floors)]
multi_level.head(10)
```

```python
multi_level = house_sales[~(price & (basement | floors))]
multi_level.head(10)
```
