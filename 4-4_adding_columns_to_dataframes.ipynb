# Chapter 4-4 Adding Columns (Series) to DataFrames

```python
import pandas as pd
```

```python
houses_path = 'data/kc_house_data.csv'
```

```python
house_sales = pd.read_csv(houses_path)
house_sales.head()
```

```python
dates = house_sales['date']
```

## Overwriting Existing Data in a Column

```python
actual_dates = pd.to_datetime(dates)
actual_dates
```

```python
house_sales['date'] = actual_dates
```

```python
house_sales.head()
```

```python
house_sales.info()
```

## Inserting data as a new column

```python
# reset the house_sales dataframe
actual_dates = pd.to_datetime(dates)
actual_dates
```

```python
actual_dates = pd.to_datetime(house_sales['date'])
```

```python
# insert column with the name "actual_dates" in the house_sales dataframe
house_sales['actual_dates'] = actual_dates 
```

```python
house_sales.head()
```

## Insert a new column in specified position

```python
# reset the house_sales dataframe
actual_dates = pd.to_datetime(dates)
actual_dates
```

```python
# create a series for dates collumn
actual_dates = pd.to_datetime(house_sales['date'])
```

```python
house_sales.insert(2, 'actual_dates', actual_dates)
```
