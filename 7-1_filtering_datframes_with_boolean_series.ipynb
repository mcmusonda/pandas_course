# Chapter 7-1: Filtering DataFrames with a Boolean Series

```python
import pandas as pd
pd.options.display.float_format = '{:.1f}'.format
```

```python
house_path = 'data/kc_house_data.csv'
```

```python
house_sales = pd.read_csv(house_path)
house_sales.head(10)
```

## Filter Houses with 3 bedrooms only
A boolean series is a pandas series where either all the values are true or false

```python
house_sales['bedrooms']
```

```python
house_sales['bedrooms'] == 3
```

```python
three_bedrooms = house_sales['bedrooms'] == 3
```

```python
house_sales[three_bedrooms]
```

```python
three_bedrooms_df = house_sales[three_bedrooms]
three_bedrooms_df.describe()
```
