```python
import pandas as pd
pd.options.display.float_format = '{:.1f}'.format
```

```python
house_path = 'data/kc_house_data.csv'
```

```python
house_sales = pd.read_csv(house_path)
house_sales.head()
```

## Equality Opposite number Inequality "!="

```python
house_sales['zipcode'] != 98198
```

```python
not_98198 = house_sales['zipcode'] != 98198
```

```python
anything_but_98198 = house_sales[not_98198]
anything_but_98198
```

```python
len(anything_but_98198)
```

```python
len(house_sales)
```

## Less than operator

```python
less_than = house_sales['price'] < 250000
less_than
```

```python
# index the dataframe
affordable_homes = house_sales[less_than]
affordable_homes.head()
```

```python
affordable_homes.describe()
```
