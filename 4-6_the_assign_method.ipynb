# Chapter 4-6 The Assign Method

```python
import pandas as pd
```

```python
houses_path = 'data/kc_house_data.csv'
```

```python
house_sales = pd.read_csv(houses_path)
house_sales.head()
```

## Using Assign method to assign a single column

```python
house_sales.assign(actual_date=pd.to_datetime(house_sales['date']))
```

```python
house_sales.head()
```

```python
house_sales.assign(date=pd.to_datetime(house_sales['date']))
```

## Adding multiple derived columns

```python
house_sales = pd.read_csv(houses_path)
house_sales.head()
```

```python
# Create series
actual_dates = pd.to_datetime(house_sales['date'])
bedrooms_plus_bathrooms = house_sales['bedrooms'] + house_sales['bathrooms']
price_per_sqft = house_sales['price'] / house_sales['sqft_living']
```

```python
augumented_house_sales = house_sales.assign(actual_dates=actual_dates, beds_plus_baths=bedrooms_plus_bathrooms, price_per_sqft=price_per_sqft)
```
