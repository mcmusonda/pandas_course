# Chapter 6-3: The loc Method

```python
import pandas as pd
pd.options.display.float_format = '{:.1f}'.format
```

```python
house_path = 'data/kc_house_data.csv'
```

```python
house_sales = pd.read_csv(house_path, index_col='id')
house_sales.head()
```

## Location specific row based on its index

```python
house_sales.loc[5631500400]
```

## Locate multiple rows based on their indices

```python
house_sales.loc[[7129300520, 5631500400, 1954400510]]
```

## Select columns to include in the result set

```python
house_sales.loc[[7129300520, 5631500400, 1954400510], ['price', 'bedrooms', 'bathrooms']]
```

## Slicing in loc method

```python
house_sales.loc[5631500400:2008000270, ['price', 'bedrooms', 'bathrooms']]
```
